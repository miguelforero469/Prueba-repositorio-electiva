```python
from matplotlib import legend
import matplotlib.pyplot  as plt
import numpy as np

# Esta es una grafica
x = np.linspace(-20,20,500)
y = x**2
plt.plot(x,y, color = "r", label = "f(x) = x^2", linewidth=2)
plt.title("Función Cuadratica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.savefig("x2.pdf")
plt.show()
```

```python
from matplotlib import legend
import matplotlib.pyplot  as plt
import numpy as np

# Esta es una grafica
x = np.linspace(-20,20,40)
y = x**2
plt.plot(x,y, ":", label = "$f(x) = x^2$", linewidth=2)
plt.title("Función Cuadratica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.savefig("x2.pdf")
plt.show()
```

```python
from matplotlib import legend
import matplotlib.pyplot  as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y, "r:", label = "$sen(x)$")
plt.plot(x,y2,"g--", label = "$cos(x)$")
plt.title("Funcion seno y coseno")
plt.xlabel("x")
plt.legend()
plt.grid()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, "r:", label = "$sen(x)$")
plt.title("Sen(x)")
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,y2,"g--", label = "$cos(x)$")
plt.title("Cos(x)")
plt.legend()
```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1,10,100)
y = np.exp(x)
y2 = np.log(x)


plt.plot(x,y,"b-", label = "$e^x$")
plt.plot(x,y2,"m--", label = "$ln(x)$")
plt.title("Funcion Exponencial y Logaritmo")
plt.xlabel("x")
plt.legend()
plt.grid()
```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5,100)
y = x**2
y2 = x**3


plt.plot(x,y,"c-", label = "$x^2$")
plt.plot(x,y2,"y--", label = "$x^3$")
plt.title("Funcion Cuadratica y Cubica")
plt.xlabel("x")
plt.legend()
plt.grid()
```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,1000)

y = np.tan(x)
y2 = 1/np.tan(x)   # cotangente


plt.plot(x,y,"b-", label = "$tan(x)$")
plt.plot(x,y2,"r--", label = "$cot(x)$")
plt.title("Funcion Tangente y Cotangente")
plt.xlabel("x")
plt.legend()
plt.grid()
```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.sqrt(x)
y2 = np.sqrt(x-2)

plt.plot(x,y,"b-", label = "$√x$")
plt.plot(x,y2,"r--", label = "$√(x-2)$")
plt.title("Funcion Raiz Cuadrada")
plt.xlabel("x")
plt.legend()
plt.grid()
```
